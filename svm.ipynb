```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn import svm
import matplotlib.pyplot as plt
```

```python
iris=datasets.load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
print(df)
x=iris.data[:,:2]
y=iris.target
model=svm.SVC(kernel='linear')
model.fit(x,y)
xx,yy=np.meshgrid(np.linspace(x[:,0].min()-1,x[:,0].max()+1,100),np.linspace(x[:,1].min()-1,x[:,1].max()+1,100))
grid_points=np.c_[xx.ravel(),yy.ravel()]
z=model.decision_function(grid_points)
print(f"shape of z = {z.shape}")
z_class_0=z[:,0]
z_class_0=z_class_0.reshape(xx.shape)
plt.contourf(xx,yy,z_class_0>0,alpha=0.8,cmap='viridis')
plt.scatter(x[:,0],x[:,1],c=y,cmap='viridis',edgecolors='k')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.show()
```
