```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
data = {
    'Age': ['<=30', '<=30', '31-40', '>40', '>40', '<=30', '31-40', '>40'],
    'Income': ['High', 'High', 'High', 'Medium', 'Low', 'Medium', 'Low', 'Medium'],
    'Student': ['No', 'Yes', 'No', 'No', 'Yes', 'Yes', 'No', 'Yes'],
    'Buys_Product': ['No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes']
}

df=pd.DataFrame(data)
df
df_encoded=pd.get_dummies(df,columns=['Age','Income','Student'],drop_first=True)
x=df_encoded.drop('Buys_Product',axis=1)
y=df_encoded['Buys_Product'].map({'No':0 , 'Yes' : 1})
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
print(f"accuracy = {model.score(x_test,y_test)}")

plt.figure(figsize=(10,8))
tree.plot_tree(model,feature_names=x.columns,class_names=['No','Yes'],filled=True)
plt.show()
```
